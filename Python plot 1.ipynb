```python
import matplotlib.pyplot as plt


x = [1,2,3,4,5,6,7,8]

y = [5,6,1,2,5,2,4,6]

plt.plot(x, y, color='blue', linestyle='dashed', linewidth = 6,
         marker='*',markerfacecolor='red', markersize=20)


plt.ylim(1,10)
plt.xlim(1,10)


plt.xlabel('Number of Students')

plt.ylabel('Subject space')


plt.title('Plot between Number of Students and Subject Space')


plt.show()



         
```
